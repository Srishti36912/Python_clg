Matplotlib is a Python library that lets you create different types of plots and charts, such as line graphs, bar charts, and scatter plots by using python scripts. It helps you visualize data clearly and effectively.

Maatplotlib is primarily used for creating 2D plots and visualizations, but it also supports 3D plotting through a specific toolkit mpl_toolkits.mplot3d module, which allows you to create 3D plots.

matplotlib.pyplot is commonly used for 2D plotting, offering easy-to-use functions to create various types of plots. With matplotlib.pyplot, you can create a wide range of 2D visualizations, including line graphs. bar charts, pie charts, and histograms.

Matplotlib does not have its own data structures. It's mainly a visualization library that works with data from other sources, like lists, NumPy arrays, and Pandas DataFrames. You provide the data, and Matplotlib helps you create plots and graphs.

# Draw a simple line graph

The plot() function is primarily used for creating line plots, which visualize data points connected by straight lines. 

```python
import matplotlib.pyplot as plt

# Draw the graph with x-coordinates [1,3] and y-coordinates [3,6]
plt.plot([1,3],[3,6])

plt.show() # Display the graph
```

As you can see, we get a straight line. The coordinates Starting Point of the line is (1,3) and the coordinates of the ending point are (3,6). You might be a little confused at this point. To understand it better, take a look at the below code.

```python
import matplotlib.pyplot as plt

x = [1,3]
y = [3,6]

plt.plot(x,y)
plt.show()
```

In the above code we just added the lists in variable namely "x" and "y". The List "x" will contain all the x-coordinates, and the list "y" will contain all the y-coordinates. Now, the coordinates of a point is determined by the numbers in the lists having same index number. Thus, we get the coordinates (1,3) and (3,6)

# Three coordinates

```python
import matplotlib.pyplot as plt

x = [1,3,6]
y = [2,10,5]

plt.plot(x,y)
plt.show()
```

# Adding multiple lines

```python
import matplotlib.pyplot as plt

x = [1,3,6]
y = [2,10,5]
x2 = [1,4,6]
y2 = [4,7,3]

plt.plot(x,y)
plt.plot(x2,y2)

plt.show()
```

As you can see, we get two different lines on the same graph with different color.

# Adding Titles, Labels, And Legend

```python
import matplotlib.pyplot as plt

x = [1,3,6]
y = [2,10,5]
x2 = [1,4,6]
y2 = [4,7,3]

plt.plot(x,y)
plt.plot(x2,y2)

plt.xlabel("X axis Name")
plt.ylabel("Y axis Name")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,3,6]
y = [2,10,5]
x2 = [1,4,6]
y2 = [4,7,3]

plt.plot(x,y, label="First Line")
plt.plot(x2,y2, label="Second Line")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() # This function will display the labels on the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,3,6]
y = [2,10,5]
x2 = [1,4,6]
y2 = [4,7,3]

plt.plot(x,y, label="First Line")
plt.plot(x2,y2, label="Second Line")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("This is matplotlib demo")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

# Plotting two lines
plt.plot(x, y1, label="Square")
plt.plot(x, y2, label="Linear")

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

Color Codes

```python
# 'b': Blue
# 'g': Green
# 'r': Red
# 'c': Cyan
# 'm': Magenta
# 'y': Yellow
# 'k': Black
# 'w': White
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.plot(x, y1,color= "b", label="Square") # color parameter allows you to specify the color of the lines or markers in a plot.
plt.plot(x, y2,color= "r", label="Linear")
plt.legend()

plt.show()
```

linestyle

```python
# '-': Solid line
# '--': Dashed line
# '-.': Dash-dot line
# ':': Dotted line
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

# Plotting with colors and line styles
plt.plot(x, y1,color= "b",linestyle='--' ,label="Square") # Blue dashed line for y1
plt.plot(x, y2,color= "r",linestyle=':', label="Linear") # Red solid line for y2

plt.legend()

plt.show()
```

Marker Style

```python
# '.': Point marker
# 'o': Circle marker
# 's': Square marker
# '^': Triangle up marker
# 'v': Triangle down marker
# '<': Triangle left marker
# '>': Triangle right marker
# 'D': Diamond marker
# 'p': Pentagon marker
# '*': Star marker
# 'h': Hexagon marker
```

```python
import matplotlib.pyplot as plt

# Data
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

# Plotting with colors, line styles and marker styles
plt.plot(x, y1, color= "b", linestyle='--' , marker='o', label="Square") # Blue dashed line with circle markers for y1
plt.plot(x, y2, color= "r", linestyle=':', marker='v', label="Linear") # Red solid line with triangle markers for y2

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```
