# Fetching rows and columns using iloc and loc

iloc and loc are two powerful ways to select and filter data from a Pandas DataFrame.

They work differently based on the type of index you want to use: integer-based or label-based.

### iloc (Integer Location):

Use Case- Select rows and columns by their position (integer index).

Works with - Integer-based positions (starting from 0).

Syntax - df.iloc[row_index, column_index]

### loc (Label Location):

Use Case- Select rows and columns by their label (row/column name).

Works with - Labels (row/column names).

Syntax - df.loc[row_label, column_label]

```python
df['winner']        # Single square brackets[]: Used to select a single column, which returns a Series.
```

```python
df['winner'].shape
```

```python
type(df['winner'])
```

```python
df[['team1','team2','winner']] # Double square brackets[[]]: Used to select a multiple columns or to ensure the result is a DataFrame,
                               #even when selecting one column
```

```python
df[['team1','team2','winner']].shape
```

```python
type(df[['team1','team2','winner']])
```

```python
df.iloc[0]
```

```python
df.iloc[1:6] # slice rows from 1 to 5
```

```python
df.iloc[0:11:2] # select rows from index 0 to 10 with a step of 2
```

```python
df.iloc[[3,55,611,129]] # fancy indexing allows you to select specific rows by providing a list of row indices, rather than a range or slice.
```

```python
df.iloc[:,[4,6,10]] # select all rows, and columns with indices 4, 6, and 10
```

```python
df.loc[3] # select the row with label
```

```python
df.loc[::2] # selects every second row, starting from the first row.
```

```python
df.loc[:,["team1","team2"]] # you cannot use integer based indexing for columns instead it requires column names (labels) to select columns.
```
