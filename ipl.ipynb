# Fetching rows and columns using iloc and loc

iloc and loc are two powerful ways to select and filter data from a Pandas DataFrame.

They work differently based on the type of index you want to use: integer-based or label-based.

### iloc (Integer Location):

Use Case- Select rows and columns by their position (integer index).

Works with - Integer-based positions (starting from 0).

Syntax - df.iloc[row_index, column_index]

### loc (Label Location):

Use Case- Select rows and columns by their label (row/column name).

Works with - Labels (row/column names).

Syntax - df.loc[row_label, column_label]

```python
# This function is designed to count the number of rows in the DataFrame where the "city" columns matches the specified city.
def get_city_match_count(city):
    filter_data = df["city"] == city
    return df[filter_data].shape[0]
```

```python
def city_match_after15(city):
    filter_data = df["city"] == city
    filter_data2 = df["season"] > 2015
    after15 = df[filter_data & filter_data2].shape[0]
    return after15
```

```python
city_match_after15("Rajkot")  # total 10 matches were played in Rajkot after 2015
```

```python
city_match_after15("Pune")  # total 11 matches were played in Pune after 2015
```

```python
city_match_after15("Indore")
```

Creating reuable functions like city_match_after15 is a great practice in programming.

It allows you to perform repetitive tasks with ease and ensures consistency in your data analysis.
